```python
# import the required python libraries like pandas and numpy

import pandas as pd
import numpy as np

# store the movie credits and movie information csv files into a dataframe into their respective variables
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")

# as we are using the overview column for recommendation
# we cannot have empty cell in the 'overview' column
# so we store only non empty cells
movies = movies[movies['overview'].notna()] 
```

```python
# to view the first 5 rows of the 'credits' dataframe
credits.head()
```

```python
# to view the first 5 rows of the 'movies' dataframe
movies.head()
```

```python
# To know the size(rows and columns) of each dataframe 
print("Credits:",credits.shape)
print("Movies Dataframe:",movies.shape)
```

```python
# renaming the column name 'movie_id' to 'id' as it is mapped to 'id' in the 'movies' dataframe
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})

# merge both the dataframe with same id
movies_merge = movies.merge(credits_column_renamed, on='id')
print(movies_merge.head())
```

```python
# as  columns'homepage', 'title_x', 'title_y', 'status','production_countries'
# are not useful to us in recommendation it is dropped
movies_cleaned = movies_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])

# let us take a look at our cleaned dataframe
print(movies_cleaned.head())
print(movies_cleaned.info())
print(movies_cleaned.head(1)['overview'])
```

```python
# using scikit-learn machine learning library 
# It features various classification, regression and clustering algorithms
# Term frequency-inverse document frequency(TF-IDF) is a text vectorizer that transforms
# the text into a usable vector which the machine/algorithm can understand
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,  
                      max_features=None,
                      strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
                      ngram_range=(1, 3),
                      stop_words = 'english')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_cleaned['overview'])
print(tfv_matrix)
print(tfv_matrix.shape)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel between 2 vectors
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_cleaned.index, index=movies_cleaned['original_title']).drop_duplicates()
print(indices)

```

```python
def give_recomendations(title, sig=sig):
    # Get the index corresponding to 'original_title' of the movie
    idx = indices[title]

    # Get the pairwsie similarity scores with entered movie
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies from highest to lowest similarity
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies is at the top
    sig_scores = sig_scores[1:11]

    # Getting their Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Display the Top 10 most similar movies with thier indices
    return movies_cleaned['original_title'].iloc[movie_indices]
```

```python
mov = input('Enter the name of a movie (present in the used dataset) : ')
print('The recommended movies are : \n')
print(give_recomendations(mov))
```
